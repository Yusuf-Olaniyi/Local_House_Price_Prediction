```python
from requests_html import HTMLSession
from bs4 import BeautifulSoup
```

```python
session = HTMLSession()
url = 'https://nigeriapropertycentre.com/for-rent'
```

```python
def getdata(url):
    r = session.get(url)
    soup = BeautifulSoup(r.text,'html.parser')
    return soup

```

```python
def getnextpage(soup):
    page = soup.find('ul',class_='pagination pagination-style-2 pagination-lg')
    while not page.find('li',class_='active disabled'):
        return 'https://nigeriapropertycentre.com' + str(page.find('li').find('a',rel = 'next')['href'])
    
```

```python
while True:
    soup = getdata('https://nigeriapropertycentre.com/for-rent')
    url = getnextpage(soup)
    if not url:
        break
    print(url)
    
```
